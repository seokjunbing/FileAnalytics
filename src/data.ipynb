```python
# df = pd.read_csv('../data/sample.tsv', sep='\t')
df = pd.read_csv(
    FILENAME, sep='\t', header=None,
    names=['filetype', 'extension', 'blocks', 'birthtime', 'alive_for_periods', 'is_dead']
)
# print(df)
# cut_buckets = {0: '0', 1: '1', 10: '10', 100: '100', 500: '500', 1000: '1000', inf: 'inf'}
cut_buckets = [0, 1, 10, 100, 500, 1000, 10000, 100000]
labels = ['%s-%s' % (cut_buckets[i - 1], cut_buckets[i]) for i in range(1, len(cut_buckets))]
# cut_buckets = {cut_buckets[i]: '%d-%d' % (cut_buckets[i - 1], cut_buckets[i]) for i in range(1, len(cut_buckets))}
cuts: pd.DataFrame= pd.cut(
    df['blocks'],
    bins=cut_buckets,
    labels=labels
)
```

```python
df['block_category'] = cuts
```
